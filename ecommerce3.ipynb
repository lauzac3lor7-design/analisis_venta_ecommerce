¿Que clientes son los mas importantes?

```python
# Los 5 de clientes que más dinero han gastado
query_clientes = """
SELECT "Customer ID", SUM(Quantity * Price) as gasto_total
FROM ventas
GROUP BY "Customer ID"
ORDER BY gasto_total DESC
LIMIT 5
"""
df_top_clientes = pd.read_sql_query(query_clientes, conn)
df_top_clientes
```

Despúes de conocer el monto total, es importante saber quienes fueron los que dieron ese dinero y asi comenzar con el estudio necesario e implementar una estrategia para que los clientes importantes siempre decidan volver y encontrar la manera de tener clientes semejantes que esten dispuestos a gastar como aquellos clientes.En caso contrario una de las consecuencia seria un impacto desproporcionado en la facturacion total en comparación con un cliente promedio

```python
query = """
WITH GastoClientes AS (
    SELECT
        "Customer ID",
        SUM(quantity * price) as gasto_individual
    FROM ventas
    GROUP BY "Customer ID"
),
GranTotal AS (
    SELECT SUM(quantity * price) as total_global FROM ventas
)
SELECT
    (SELECT SUM(gasto_individual) FROM (SELECT gasto_individual FROM GastoClientes ORDER BY gasto_individual DESC LIMIT 10)) as gasto_top_10,
    total_global
FROM GranTotal;
"""

df_riesgo = pd.read_sql_query(query, conn)

# Cálculos de porcentaje
gasto_top = df_riesgo['gasto_top_10'][0]
total_global = df_riesgo['total_global'][0]
porcentaje_top = (gasto_top / total_global) * 100

print(f"📊 Análisis de Concentración:")
print(f"--------------------------------")
print(f"💰 Gasto del Top 10 Clientes: ${gasto_top:,.2f}")
print(f"🌍 Ingreso Total Global:      ${total_global:,.2f}")
print(f"🚀 Los 10 mejores clientes representan el {porcentaje_top:.2f}% de tus ventas totales.")
```

Ahora ya se tiene un porcentaje que mostrar para saber que el 16.25% es la representacion de las ventas totales de los mejores 10 clientes. Tener ese porcentaje en solo 10 clientes indica una dependencia moderada

```python
query = """
SELECT
    CASE strftime('%w', invoicedate)
        WHEN '0' THEN '0. Domingo'
        WHEN '1' THEN '1. Lunes'
        WHEN '2' THEN '2. Martes'
        WHEN '3' THEN '3. Miércoles'
        WHEN '4' THEN '4. Jueves'
        WHEN '5' THEN '5. Viernes'
        WHEN '6' THEN '6. Sábado'
    END AS dia_semana,
    COUNT(DISTINCT invoice) AS total_pedidos,
    ROUND(SUM(quantity * price), 2) AS ingresos_totales
FROM ventas
GROUP BY dia_semana
ORDER BY ingresos_totales DESC;
"""

df_resultados = pd.read_sql_query(query, conn)

df_resultados
```

En esta tabla se puede ver el comportamiento semanal.Los jueves son los dias donde hay una mayor venta generando £3.6 millones en ingresos con 9,710 pedidos, por el contario el sabado se muestra una caida drástica con solo £9,803 y esto podría ser porque la tienda no opera regularmente, ya que el domingo vuelve a recuperarse segun los ingresos mostrados.

```python
import matplotlib.pyplot as plt



query = """
SELECT
    strftime('%m', invoicedate) AS mes_num,
    CASE strftime('%m', invoicedate)
        WHEN '01' THEN 'Enero' WHEN '02' THEN 'Febrero' WHEN '03' THEN 'Marzo'
        WHEN '04' THEN 'Abril' WHEN '05' THEN 'Mayo' WHEN '06' THEN 'Junio'
        WHEN '07' THEN 'Julio' WHEN '08' THEN 'Agosto' WHEN '09' THEN 'Septiembre'
        WHEN '10' THEN 'Octubre' WHEN '11' THEN 'Noviembre' WHEN '12' THEN 'Diciembre'
    END AS mes_nombre,
    ROUND(SUM(quantity * price), 2) AS ingresos_mensuales
FROM ventas
WHERE mes_nombre IS NOT NULL  -- Esto evita el error de Matplotlib
GROUP BY mes_num
ORDER BY mes_num ASC;
"""

df_estacionalidad = pd.read_sql_query(query, conn)


# Grafica
plt.figure(figsize=(12,6))
# Convertimos a string explícitamente para asegurar que Matplotlib lo entienda
plt.plot(df_estacionalidad['mes_nombre'].astype(str),
         df_estacionalidad['ingresos_mensuales'],
         marker='o', color='royalblue', linewidth=2)

plt.title('Tendencia de Ventas por Mes (Análisis de Estacionalidad)', fontsize=15)
plt.xlabel('Mes')
plt.ylabel('Ingresos ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Al observar la tendencia de ventas por mes, se logra ver un crecimiento exponencial que inicia desde septiembre y teniendo sus mejores ventas durante los ultimos meses pero Noviembre liderando, se creeria que es diciembre debido a que es un mes lleno de festividades, no solo para Reino Unido si para todos los paises que compran en esta tienda, por eso fue importante analizar mas de fondo esre mes con el comportamiento de ventas Diario

```python
import matplotlib.pyplot as plt

# Consulta SQL para Noviembre (Mes '11')
query_noviembre = """
SELECT
    strftime('%d', invoicedate) AS dia,
    ROUND(SUM(quantity * price), 2) AS ventas_diarias
FROM ventas
WHERE strftime('%m', invoicedate) = '11'
GROUP BY dia
ORDER BY dia ASC;
"""

df_noviembre = pd.read_sql_query(query_noviembre, conn)

# Limpieza de posibles valores nulos
df_noviembre = df_noviembre.dropna()

plt.figure(figsize=(14, 6))
plt.plot(df_noviembre['dia'], df_noviembre['ventas_diarias'],
         marker='o', color='darkorange', linewidth=2, label='Ventas diarias (Noviembre)')

plt.title('Zoom a Noviembre: Comportamiento de Ventas Diario', fontsize=16)
plt.xlabel('Día de Noviembre')
plt.ylabel('Ingresos (£)')
plt.xticks(df_noviembre['dia'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.fill_between(df_noviembre['dia'], df_noviembre['ventas_diarias'], color='darkorange', alpha=0.1)
plt.legend()
plt.show()
```

Con esta grafica se puede decir que el jueves es el dia de mayor actividad con £3.6 millones y el mayor volumen de pedidos con 9,710. Gran parte de los cliente son los pequeños negocios o tiendas locales que compran al mayor y suelen hacerlo este dia para asegurar que el inventario llegue el lunes o el viernes y la tienda tiene tiempo de preparar el paquete y enviarlo antes de el cierre del viernes.

```python
import seaborn as sns
query = """
SELECT
    stockcode,
    description,
    AVG(price) AS precio_promedio,
    SUM(quantity) AS cantidad_total
FROM ventas
GROUP BY stockcode
HAVING cantidad_total > 0 AND precio_promedio > 0
LIMIT 500;  -- Limitamos para una visualización clara
"""

df_relacion = pd.read_sql_query(query, conn)


# Calculamos la correlación numérica
correlacion = df_relacion['precio_promedio'].corr(df_relacion['cantidad_total'])

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_relacion, x='precio_promedio', y='cantidad_total', alpha=0.6)
plt.title(f'Relación Precio vs Cantidad (Correlación: {correlacion:.2f})')
plt.xlabel('Precio Unitario ($)')
plt.ylabel('Cantidad Total Vendida')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

print(f"El coeficiente de correlación es: {correlacion:.2f}")
```

Esta gráfica muestra la relación entre el precio Unitario y la cantidad total vendida, esto es útil para entender las estrategias de los precios de negocio. Se obtiene un coeficiente de correlación de -0.11 que indica una relación negativa muy ligera. Esto confirma una regla del mercado: a medida que el precio aumenta, la cantidad vendida, pero de forma muy gradual, es decir que los clientes no son extremadamente sensibles al precio en todos los rangos o que la necesidad del producto supera al factor costo.

```python
query = """
WITH ComprasPorCliente AS (
    SELECT
        "Customer ID",
        description,
        SUM(quantity) as cantidad_total,
        ROUND(SUM(quantity * price), 2) as gasto_total
    FROM ventas
    GROUP BY "Customer ID", description
)
SELECT "Customer ID", description, cantidad_total, gasto_total
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY "Customer ID" ORDER BY gasto_total DESC) as ranking
    FROM ComprasPorCliente
)
WHERE ranking <= 3
ORDER BY "Customer ID", ranking;
"""

df_afinidad = pd.read_sql_query(query, conn)


df_afinidad.head(10)
```

```python
query = """
SELECT
    country,
    Price,
    Quantity,
    (Price * Quantity) as total_venta,
    strftime('%H', invoicedate) as hora_dia
FROM ventas
"""
df_influencia = pd.read_sql_query(query, conn)

# 2. Análisis de Correlación (Para ver qué variable se mueve más con el éxito)
# Convertimos la hora a número para poder compararla
df_influencia['hora_dia'] = df_influencia['hora_dia'].astype(int)
correlaciones = df_influencia[['Price', 'Quantity', 'hora_dia', 'total_venta']].corr()['total_venta'].sort_values(ascending=False)


plt.figure(figsize=(10, 5))
correlaciones.drop('total_venta').plot(kind='bar', color='teal')
plt.title('¿Qué factor influye más en el monto de la venta?')
plt.ylabel('Nivel de Influencia (Correlación)')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()

print("Factores de influencia ordenados:")
print(correlaciones.drop('total_venta'))
```

Este gráfico de barras mide la correlación entre distintos factores y el monto total de cada venta.La variable quantity tiene la correlacíón mas alta con 85.4% y significa que el monto de la factura depende mucho mas de cuántas unidades compra el cliente que de cualquier otra cosa.
