Despúes de conocer el monto total, es importante saber quienes fueron los que dieron ese dinero y asi comenzar con el estudio necesario e implementar una estrategia para que los clientes importantes siempre decidan volver y encontrar la manera de tener clientes semejantes que esten dispuestos a gastar como aquellos clientes.En caso contrario una de las consecuencia seria un impacto desproporcionado en la facturacion total en comparación con un cliente promedio

```python
query = """
WITH GastoClientes AS (
    SELECT
        "Customer ID",
        SUM(quantity * price) as gasto_individual
    FROM ventas
    GROUP BY "Customer ID"
),
GranTotal AS (
    SELECT SUM(quantity * price) as total_global FROM ventas
)
SELECT
    (SELECT SUM(gasto_individual) FROM (SELECT gasto_individual FROM GastoClientes ORDER BY gasto_individual DESC LIMIT 10)) as gasto_top_10,
    total_global
FROM GranTotal;
"""

df_riesgo = pd.read_sql_query(query, conn)

# Cálculos de porcentaje
gasto_top = df_riesgo['gasto_top_10'][0]
total_global = df_riesgo['total_global'][0]
porcentaje_top = (gasto_top / total_global) * 100

print(f"📊 Análisis de Concentración:")
print(f"--------------------------------")
print(f"💰 Gasto del Top 10 Clientes: £{gasto_top:,.2f}")
print(f"🌍 Ingreso Total Global:      £{total_global:,.2f}")
print(f"🚀 Los 10 mejores clientes representan el {porcentaje_top:.2f}% de tus ventas totales.")
```

Ahora ya se tiene un porcentaje que mostrar para saber que el 16.04% es la representacion de las ventas totales de los mejores 10 clientes. Tener ese porcentaje en solo 10 clientes indica una dependencia moderada

```python
query = """
SELECT
    CASE strftime('%w', invoicedate)
        WHEN '0' THEN '0. Domingo'
        WHEN '1' THEN '1. Lunes'
        WHEN '2' THEN '2. Martes'
        WHEN '3' THEN '3. Miércoles'
        WHEN '4' THEN '4. Jueves'
        WHEN '5' THEN '5. Viernes'
        WHEN '6' THEN '6. Sábado'
    END AS dia_semana,
    COUNT(DISTINCT invoice) AS total_pedidos,
    ROUND(SUM(quantity * price), 2) AS ingresos_totales
FROM ventas
GROUP BY dia_semana
ORDER BY ingresos_totales DESC;
"""

df_resultados = pd.read_sql_query(query, conn)

df_resultados
```

```python
import matplotlib.pyplot as plt



query = """
SELECT
    strftime('%Y-%m', invoicedate) AS mes_num,
    (strftime('%Y', invoicedate) || '-' ||
     CASE strftime('%m', invoicedate)
        WHEN '01' THEN 'Enero' WHEN '02' THEN 'Febrero' WHEN '03' THEN 'Marzo'
        WHEN '04' THEN 'Abril' WHEN '05' THEN 'Mayo' WHEN '06' THEN 'Junio'
        WHEN '07' THEN 'Julio' WHEN '08' THEN 'Agosto' WHEN '09' THEN 'Septiembre'
        WHEN '10' THEN 'Octubre' WHEN '11' THEN 'Noviembre' WHEN '12' THEN 'Diciembre'
     END) AS mes_nombre,
    ROUND(SUM(quantity * price), 2) AS ingresos_mensuales
FROM ventas
WHERE invoicedate IS NOT NULL
GROUP BY mes_num
ORDER BY mes_num ASC;
"""


df_estacionalidad = pd.read_sql_query(query, conn)


# Grafica
plt.figure(figsize=(12,6))
# Convertimos a string explícitamente para asegurar que Matplotlib lo entienda
plt.plot(df_estacionalidad['mes_nombre'].astype(str),
         df_estacionalidad['ingresos_mensuales'],
         marker='o', color='royalblue', linewidth=2)

plt.title('Tendencia de Ventas por Mes (Análisis de Estacionalidad)', fontsize=15)
plt.xlabel('Mes')
plt.ylabel('Ingresos (£)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

En esta gráfica se muestra los ingresos mensuales del 2009 al 2011. En el mes de enero a agosto las ventas se mantienen en un rango relativamente constante de aproximadamente de 0.5 M a 0.65 M.Se ve un crecimiento muy marcado desde septiembre con picos fuertes en octubre-noviembre llegando cerca de 1.0M-1.1M.

```python
import matplotlib.pyplot as plt

# Consulta SQL para Noviembre (Mes '11')
query_noviembre = """
SELECT
    date(invoicedate) AS dia,
    ROUND(SUM(quantity * price), 2) AS ventas_diarias
FROM ventas
WHERE strftime('%m', invoicedate) = '11'
GROUP BY dia
ORDER BY dia ASC;
"""


df_noviembre = pd.read_sql_query(query_noviembre, conn)

# Limpieza de posibles valores nulos
df_noviembre = df_noviembre.dropna()

df_noviembre['dia'] = pd.to_datetime(df_noviembre['dia'])

df_noviembre['anio'] = df_noviembre['dia'].dt.year
df_noviembre['dia_mes'] = df_noviembre['dia'].dt.day

plt.figure(figsize=(14, 6))
for anio, g in df_noviembre.groupby('anio'):
    g = g.sort_values('dia_mes')
    plt.plot(g['dia_mes'], g['ventas_diarias'], marker='o', linewidth=2, label=f'Noviembre {anio}')

plt.title('Zoom a Noviembre: Comportamiento de Ventas Diario', fontsize=16)
plt.xlabel('Día del mes')
plt.ylabel('Ingresos (£)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

Con esta grafica se puede visualizar el comportamiento de diario en el mes de noviembre.En el eje x se observan fechas de noviembre en mas de un año del 2010 y 2011, y en el eje Y los ingresos diarios. Se ven fluctuaciones es decir dias normales en un rango de que van de 30k y 50k y picos que pueden llegar a 70k y 80k, asi que el negocio no vende de una manera uniforme y hay dias donde entran pedidos grandes o muchos pedidos a la vez, como hay picos en dos bloques noviembre 2010 y noviembre 2011, en mas de un año, noviembre presenta dias con picos relevantes lo cual refuerza que es un mes con estacionalidad.

¿Los picos ocurren por más pedidos o por pedidos más grandes?

```python
query_nov_detalle = """
SELECT
  date(invoicedate) AS fecha,
  COUNT(DISTINCT invoice) AS pedidos,
  ROUND(SUM(quantity * price), 2) AS ingresos,
  ROUND(SUM(quantity * price) * 1.0 / COUNT(DISTINCT invoice), 2) AS ticket_promedio
FROM ventas
WHERE strftime('%m', invoicedate) = '11'
GROUP BY fecha
ORDER BY fecha;
"""
df_nov_detalle = pd.read_sql_query(query_nov_detalle, conn)
df_nov_detalle['fecha'] = pd.to_datetime(df_nov_detalle['fecha'])

df_nov_detalle.sort_values('ingresos', ascending=False).head(5)
```

Esta tabla esta separada por peididos y ticket promedio, esta dividida por fecha especifica, hay dos picos 2010-11-04 con 126 pedidos y ticket £665 con ingresos de £83,876 y 2010-11-25 con 146 pedidos pero ticket £464 y ingresos de£67,870, aunque el 25/11 tuvo más pedidos, el 04/11 generó más ingresos porque el ticket promedio fue mucho mayor. En dias pico el tamaño del pedido puede pesar tanto o más que el número de pedidos.

```python
pedidos = df_nov_detalle['pedidos'].median()
ticket  = df_nov_detalle['ticket_promedio'].median()

top5 = df_nov_detalle.sort_values('ingresos', ascending=False).head(5).copy()

top5['pedidos_%'] = (top5['pedidos'] / pedidos - 1) * 100
top5['ticket_%']  = (top5['ticket_promedio'] / ticket - 1) * 100

top5[['fecha','pedidos','ticket_promedio','ingresos','pedidos_%','ticket_%']]
```

Los picos de facturación en noviembre se explican por una combinación de más pedidos y pedidos de mayor tamaño. En los 5 días de mayores ingresos, los pedidos aumentan aproximadamente entre +26% y +48% respecto al día típico, y el ticket promedio entre +14% y +45%, lo que se traduce en días con ingresos entre 68% y 107% superiores a un día promedio. Esto indica que, para capturar al máximo la temporada alta, es clave optimizar tanto la capacidad operativa para procesar más pedidos como las estrategias para incrementar el ticket.

```python
import seaborn as sns
query = """
SELECT
    stockcode,
    description,
    AVG(price) AS precio_promedio,
    SUM(quantity) AS cantidad_total
FROM ventas
GROUP BY stockcode
HAVING cantidad_total > 0 AND precio_promedio > 0;
"""

df_relacion = pd.read_sql_query(query, conn)


# Calculamos la correlación numérica
correlacion = df_relacion['precio_promedio'].corr(df_relacion['cantidad_total'])

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_relacion, x='precio_promedio', y='cantidad_total', alpha=0.6)
plt.title(f'Relación Precio vs Cantidad (Correlación: {correlacion:.2f})')
plt.xlabel('Precio Unitario (£)')
plt.ylabel('Cantidad Total Vendida')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

print(f"El coeficiente de correlación es: {correlacion:.2f}")
```

Esta gráfica muestra la relación entre el precio Unitario y la cantidad total vendida, esto es útil para entender las estrategias de los precios de negocio. Se obtiene un coeficiente de correlación de -0.06 que indica una relación negativa muy ligera. Esto confirma una regla del mercado: a medida que el precio aumenta, la cantidad vendida, pero de forma muy gradual, es decir que los clientes no son extremadamente sensibles al precio en todos los rangos o que la necesidad del producto supera al factor costo.

```python
query = """
WITH ComprasPorCliente AS (
    SELECT
        "Customer ID",
        description,
        SUM(quantity) as cantidad_total,
        ROUND(SUM(quantity * price), 2) as gasto_total
    FROM ventas
    GROUP BY "Customer ID", description
)
SELECT "Customer ID", description, cantidad_total, gasto_total
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY "Customer ID" ORDER BY gasto_total DESC) as ranking
    FROM ComprasPorCliente
)
WHERE ranking <= 3
ORDER BY "Customer ID", ranking;
"""

df_afinidad = pd.read_sql_query(query, conn)


df_afinidad.head(10)
```

```python
query = """
SELECT
    invoice,
    country,
    SUM(Price * Quantity) as total_venta,
    SUM(Quantity) as total_unidades,
    AVG(Price) as precio_promedio,
    strftime('%H', invoicedate) as hora_dia
FROM ventas
GROUP BY invoice, country, hora_dia
"""
df_influencia = pd.read_sql_query(query, conn)

# 2. Análisis de Correlación (Para ver qué variable se mueve más con el éxito)
# Convertimos la hora a número para poder compararla
df_influencia['hora_dia'] = df_influencia['hora_dia'].astype(int)
correlaciones = df_influencia[['precio_promedio', 'total_unidades', 'hora_dia', 'total_venta']].corr()['total_venta'].sort_values(ascending=False)

plt.figure(figsize=(10, 5))
correlaciones.drop('total_venta').plot(kind='bar', color='teal')
plt.title('¿Qué factor influye más en el monto de la venta?')
plt.ylabel('Nivel de Influencia (Correlación)')
plt.axhline(0, color='black', linewidth=0.8)
plt.show()

print("Factores de influencia ordenados:")
print(correlaciones.drop('total_venta'))
# Cerramos la conexión a la base de datos
conn.close()
```

Este gráfico de barras mide la correlación entre distintos factores por medio de la correlación de Pearson entre el monto total de la venta y los tres factores.Los valores cercanos a 1 indican relación positiva, cercanos a -1 relacion negativa y cercano a 0 significa poca o ninguna relación.El factor que está más asociado al monto total es el volumen, es decir total_unidades.Por lo tanto, el crecimiento en ingresos se impulsa principalmente por vender mas unidades por compra.
